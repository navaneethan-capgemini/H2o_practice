```python
factorsList = ["X6: Orientation", "X8: Glazing area distribution"]
df[factorsList] = df[factorsList].asfactor() # Converting these columns as enums or factors
#df_h2o=h2o.H2OFrame(df)
#df.types
```

```python
train, test = df.split_frame([0.8]) # splitting into train,test
```

```python
x=df.names
y="Y2: Cooling Load" # The response column 
```

```python
train.as_data_frame().hist(figsize=(20,20))
plt.show() #visualising the bell curve for factors
```

```python
res = train[x].cor(train[y]).as_data_frame()
res.index = x
res.plot.barh()
plt.show() # plotting the pearson's correlation w.r.t Y2 
```
