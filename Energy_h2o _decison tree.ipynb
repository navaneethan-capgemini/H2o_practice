```python
#import sys
import h2o
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
h2o.init()
```

```python
df=h2o.import_file('F:\H2o\ENB2012_data.csv') #importing data
```

```python
df.describe()#describing data
```

```python
df.as_data_frame() #printing data
```

```python
df.names
```

```python
factorsList = ["X6: Orientation", "X8: Glazing area distribution"]
df[factorsList] = df[factorsList].asfactor() # Converting these columns as enums or factors
#df_h2o=h2o.H2OFrame(df)
#df.types
```

```python
train, test = df.split_frame([0.8]) # splitting into train,test
```

```python
x=df.names
y="Y2: Cooling Load" # The response column 
```

```python
train.as_data_frame().hist(figsize=(20,20))
plt.show() #visualising the bell curve for factors
```

```python
res = train[x].cor(train[y]).as_data_frame()
res.index = x
res.plot.barh()
plt.show() # plotting the pearson's correlation w.r.t Y2 
```

```python
train.cor().round(2) # printing the pearson's correlarion for all factors
```

```python
#importing decision tree and training it

m = h2o.estimators.H2ORandomForestEstimator(model_id="RF_defaults", nfolds=10)
m.train(x, y, train)
```

```python
m #printing the model
```

```python
m.model_performance(test)
```

```python
pred=m.predict(test)
```

```python
m.model_performance(test)
```
