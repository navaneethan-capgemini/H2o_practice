```python
from platform import python_version
print(python_version())
import h2o
```

```python
h2o.init() #h2o intialization
```

```python
df = h2o.import_file("F:\H2o\Iris.csv") #importing dataset from local
```

```python
#print(df)
data=df.as_data_frame() # printing data
data
```

```python
y="Species" # The response column or dependant variable
x=df.names #The independant variables or factors
x.remove(y) #Removing y from independant variables
print(x) # printing the independant variables
```

```python
train, test = df.split_frame([0.8]) # Splitting the data into test and test
```

```python
train.as_data_frame()
```

```python
from h2o.estimators.deeplearning import H2ODeepLearningEstimator # selecting deepLearning estimatatormodel
m = H2ODeepLearningEstimator()
m.train(x, y, train) # Training the model
```

```python
m # printing the model details
```

```python
p=m.predict(test) # predicting using test
p.as_data_frame() # printing the prediction
```

```python
(p["predict"]== test["Species"]).mean() # comparing actual vs prediction and finding the accuracy 
```

```python
p["predict"].cbind(test["Species"]).as_data_frame() # Comparing the predicted vs actual values
```

```python
m.model_performance(test) # Model performance of test
```
